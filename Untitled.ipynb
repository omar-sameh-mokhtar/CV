```python
import cv2 as cv
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def decode_barcode_image(processed_image):
    """Decode Code 11 barcode from the provided processed binary image."""
    
    # Define bar encoding for Code 11
    NARROW_BAR = "0"
    WIDE_BAR = "1"
    code11_mappings = {
        "00110": "Start/Stop",
        "10001": "1",
        "01001": "2",
        "11000": "3",
        "00101": "4",
        "10100": "5",
        "01100": "6",
        "00011": "7",
        "10010": "8",
        "10000": "9",
        "00001": "0",
        "00100": "-",
    }

    # Calculate the column-wise mean of the barcode image and threshold to binary
    try:
        inverted_image = 255 - processed_image  # Manually invert the image
        column_means = inverted_image.mean(axis=0)  # Calculate the mean along the columns
        binary_columns = (column_means <= 127).astype(np.uint8)
    except Exception as e:
        print("Error processing column means or binary columns:", e)
        return []

    # Convert the binary columns to a string for decoding
    pixel_string = ''.join(binary_columns.astype(np.uint8).astype(str))

    # Determine the size of a narrow bar
    try:
        narrow_bar_size = 0
        for pixel in pixel_string:
            if pixel == "1":
                narrow_bar_size += 1
            else:
                break
        wide_bar_size = narrow_bar_size * 2  # Wide bar is twice as large as a narrow bar
    except Exception as e:
        print("Error calculating bar sizes:", e)
        return []

    # Decode the barcode by identifying narrow and wide bars
    decoded_characters = []
    current_index = 0
    current_bar_widths = ""
    skip_next_separator = False

    while current_index < len(pixel_string):
        if skip_next_separator:
            current_index += narrow_bar_size
            skip_next_separator = False
            continue

        # Count consecutive pixels with the same value
        run_length = 1
        try:
            while (
                current_index + run_length < len(pixel_string)
                and pixel_string[current_index] == pixel_string[current_index + run_length]
            ):
                run_length += 1
        except Exception:
            pass

        current_index += run_length

        # Determine if the bar is narrow or wide
        if run_length == narrow_bar_size:
            current_bar_widths += NARROW_BAR
        elif run_length == wide_bar_size:
            current_bar_widths += WIDE_BAR

        # Check if the current pattern matches a Code 11 character
        if current_bar_widths in code11_mappings:
            decoded_characters.append(code11_mappings[current_bar_widths])
            current_bar_widths = ""
            skip_next_separator = True  

    print(f"The final decoded barcode is: {decoded_characters}")
    return decoded_characters
```

```python
def isolate_barcode_section(image):
    """ Detect and Extract Barcode Regions from the Image. """
    
    # Detect outlines in th image
    outlines, _ = cv.findContours(image, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

    if not outlines:
        print("No outlines detected.")
        return None, image

    # Filter outlines based on relative area
    primary_outline = max(outlines, key=cv.contourArea)
    primary_area = cv.contourArea(primary_outline)

    relevant_outlines = [
        outline for outline in outlines if cv.contourArea(outline) >= 0.05 * primary_area
    ]

    if not relevant_outlines:
        print("No relevant outlines detected.")
        return None, image

    # Calculate a bounding rectangle around all relevant outlines
    all_points = np.vstack(relevant_outlines) 
    x, y, w, h = cv.boundingRect(all_points)

    # Crop the image to show barcode only
    barcode_only = image[y:y + h, x:x + w]

    # Visualize the detected region
    outlined_image = cv.cvtColor(cv.bitwise_not(image), cv.COLOR_GRAY2BGR) 
    cv.rectangle(outlined_image, (x, y), (x + w, y + h), (0, 255, 0), 4)  

    return barcode_only, outlined_image
```

```python
def visualize_image_with_title(image, title):
    """Display the image with a border and title."""
    if image is None:
        print("No image to display.")
        return

    # Create a border around the image
    bordered_image = cv.copyMakeBorder(
        image, top=10, bottom=10, left=10, right=10,
        borderType=cv.BORDER_CONSTANT, value=(255, 0, 0)  # Red border
    )

    # Convert the image to RGB for correct visualization
    if len(image.shape) == 3:  # Check if the image is colored
        rgb_image = cv.cvtColor(bordered_image, cv.COLOR_BGR2RGB)
    else:  # Grayscale
        rgb_image = bordered_image

    plt.figure(figsize=(8, 8))
    plt.imshow(rgb_image, cmap='gray' if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def extract_grayscale_regions_hsv(input_img, saturation_threshold=30):
    """ Extract grayscale regions from an image using low saturation in the HSV color space."""
    # Convert the image to HSV color space
    hsv_img = cv.cvtColor(input_img, cv.COLOR_BGR2HSV)
    
    # Split the HSV channels
    h, s, v = cv.split(hsv_img)
    
    # Create a mask for low-saturation regions
    low_saturation_mask = s < saturation_threshold
    
    # Convert the mask to binary
    binary_mask = low_saturation_mask.astype(np.uint8) * 255
    
    # Create a white background
    white_background = np.ones_like(input_img, dtype=np.uint8) * 255 
    
    # Use the mask to isolate low-saturation regions (grayscale)
    grayscale_regions = cv.bitwise_and(input_img, input_img, mask=binary_mask)
    
    # Invert the mask
    inverted_mask = cv.bitwise_not(binary_mask)
    grayscale_regions += cv.bitwise_and(white_background, white_background, mask=inverted_mask)
    
    return grayscale_regions


```

```python
def process_barcode_with_thresholding(input_image):
    """Process the barcode area using vertical smoothing followed by Black and White conversion."""
    _, black_and_white = cv.threshold(cv.cvtColor(input_image, cv.COLOR_BGR2GRAY), 0, 255, cv.THRESH_BINARY_INV | cv.THRESH_OTSU)
    visualize_image_with_title(black_and_white, ' black_and_white_image ')
    return black_and_white
```

```python
def process_barcode_area_closing(black_and_white):
    """Process the barcode area using morphological operations."""
    
    # Define a vertical kernel for morphological operations
    vertical_structure_kernel = cv.getStructuringElement(cv.MORPH_RECT, (1, 50))
    
    # Perform morphological opening to emphasize vertical structures
    processed_image = cv.erode(black_and_white, vertical_structure_kernel, iterations=1)
    processed_image = cv.dilate(processed_image, vertical_structure_kernel, iterations=1)  # Equivalent to MORPH_OPEN
    
    # Display the result of the morphological openning
    visualize_image_with_title(processed_image, 'Processed Image (Morphological Opening)')
    
    # Crop the barcode area using outlines
    barcode_segment, contour_visualization = isolate_barcode_section(processed_image)
    visualize_image_with_title(cv.bitwise_not(barcode_segment), 'Segmented Barcode Area')

    # Apply morphological closing to fill small gaps with the kernel size
    closing_kernel = cv.getStructuringElement(cv.MORPH_RECT, (1, 120))  
    dilated_image = cv.dilate(barcode_segment, closing_kernel, iterations=3)
    gap_filled_image = cv.erode(dilated_image, closing_kernel, iterations=3)  
    
    visualize_image_with_title(cv.bitwise_not(gap_filled_image), 'Final Barcode Area')

    return gap_filled_image
```

```python
import numpy as np
import cv2 as cv
import numpy as np

def decode_barcode_from_image(processed_image):
    """Decode Code 11 barcode from a provided image (as a NumPy array)."""
    # Define Code 11 encoding
    NARROW_BAR = "0"
    WIDE_BAR = "1"
    code11_mappings = {
        "00110": "Start/Stop",
        "10001": "1",
        "01001": "2",
        "11000": "3",
        "00101": "4",
        "10100": "5",
        "01100": "6",
        "00011": "7",
        "10010": "8",
        "10000": "9",
        "00001": "0",
        "00100": "-",
    }

    # Ensure the image is in binary format (thresholding)
    _, binary_image = cv2.threshold(processed_image, 127, 255, cv2.THRESH_BINARY_INV)

    # Calculate the column-wise mean and convert it to binary
    column_means = binary_image.mean(axis=0)
    binary_columns = (column_means > 127).astype(np.uint8)  # Black = 1, White = 0

    # Convert the binary columns to a string for decoding
    pixel_string = ''.join(binary_columns.astype(str))

    # Determine narrow and wide bars dynamically
    narrow_bar_size = 1  # Single pixel as a narrow bar
    wide_bar_size = 2 * narrow_bar_size  # Wide bar twice the size of a narrow bar

    # Decode the barcode by identifying narrow and wide bars
    decoded_characters = []
    current_index = 0
    current_bar_widths = ""
    skip_next_separator = False

    while current_index < len(pixel_string):
        if skip_next_separator:
            current_index += narrow_bar_size
            skip_next_separator = False
            continue

        # Count consecutive pixels with the same value
        run_length = 1
        while (
            current_index + run_length < len(pixel_string)
            and pixel_string[current_index] == pixel_string[current_index + run_length]
        ):
            run_length += 1

        current_index += run_length

        # Classify as narrow or wide bar
        if run_length <= wide_bar_size:
            current_bar_widths += NARROW_BAR
        else:
            current_bar_widths += WIDE_BAR

        # Match the pattern to Code 11
        if current_bar_widths in code11_mappings:
            decoded_characters.append(code11_mappings[current_bar_widths])
            current_bar_widths = ""
            skip_next_separator = True

    print(f"Decoded barcode: {''.join(decoded_characters)}")
    return decoded_characters



def detectBarcode(img):
    distances, maxdistance = detect_distance_between_lines(img)
    print(maxdistance)
    kernel = cv.getStructuringElement(cv.MORPH_RECT, (maxdistance+1, 1)) #+1 3shan maxdistance is 1 pixel short
    closed_image = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

    _, closed_image_2 = cv.threshold(closed_image, 150, 255, cv.THRESH_BINARY_INV)
    #cv.imshow('imagle', closed_image_2)
    #cv.waitKey(0)
    contours, _ = cv.findContours(closed_image_2, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    largest_contour = None
    max_area = 0

    for contour in contours:
        area = cv.contourArea(contour)
        if area > max_area:
            max_area = area
            largest_contour = contour

    x, y, deltax, deltay = cv.boundingRect(largest_contour)
    #print(x,y,deltax,deltay)
    mabrook = img[y:y+deltay, x:x+deltax]
    return mabrook
def detect_distance_between_lines(gray):

    _, binary_image = cv.threshold(gray, 150, 255, cv.THRESH_BINARY_INV)  

    edges = cv.Canny(binary_image, 50, 150, apertureSize=3)
    
    lines = cv.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=100, maxLineGap=10)

    line_image = np.copy(gray)

    uniform = 1 #needed when rotating only

    if lines is not None:
        # Sort lines by x
        lines = sorted(lines, key=lambda x: x[0][0])

        distances = []
        maxdistance = 0

        for i in range(1, len(lines)):
            # Get the x of the start and end points of  lines
            x1, y1, x2, y2 = lines[i-1][0] #2 point le awl line w 2 le tany line
            x3, y3, x4, y4 = lines[i][0]

            if abs(x3 - x1) == abs(x4 - x2): #to ignore some irregular lines
                distance = abs(x3 - x1)
                if distance > maxdistance:
                    maxdistance = distance
                distances.append(distance)
            
            cv.line(line_image, (x1, y1), (x2, y2), 1)  
            cv.line(line_image, (x3, y3), (x4, y4), 1)
        #cv.imshow('Detected Lines', line_image)
        #cv.waitKey(0)

        if len(distances) < 50: #all 
            uniform = 0

        print(f"Distances between barcode bars: {distances}")
        return uniform, maxdistance

    else:
        print("No lines detected.")
        return uniform, -1
def detect_rotation(image): 
    
    edges = cv.Canny(image, 50, 150, apertureSize=3)
    
    lines = cv.HoughLines(edges, 1, np.pi / 180, 150)
    
    angles = []
    if lines is not None:
        for rho, theta in lines[:, 0]:
            angle = np.degrees(theta) - 90  # Convert theta to degrees and normalize
            angles.append(angle)
    
    # Calculate the dominant angle (median)
    if angles:
        dominant_angle = np.median(angles)
    else:
        dominant_angle = 404  
    print(f"The detected rotation angle is approximately {dominant_angle:.2f} degrees.")
    return int(dominant_angle) != -90 and int(dominant_angle) != 0

def rotate(img, angle, rotPoint=None):
    (height, width) = img.shape[:2] #first 2 values

    if rotPoint is None:
        rotPoint = (width//2, height//2)
    
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0) #1.0 is rescale parameter
    dimensions = (width, height)

    return cv.warpAffine(image, rotMat, dimensions, flags=cv.INTER_CUBIC, borderMode=cv.BORDER_REPLICATE)

def Handle_rotation(image):
    display(image,"s")
    imageP = cv.copyMakeBorder(image, 50, 50, 50, 50, cv.BORDER_CONSTANT, value=(255, 255, 255))
    edges = cv.Canny(imageP, 50, 200)
    contours, _ = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    largest_contour = max(contours, key=cv.contourArea)
    rect = cv.minAreaRect(largest_contour)

    angle = rect[2]

    (h, w) = imageP.shape[:2]
    center = (w // 2, h // 2)

    isValid, _ = detect_distance_between_lines(image)
    print (isValid)
    if isValid == 0:
        angle-=90

    M = cv.getRotationMatrix2D(center, angle, 1.0)
    
    image = cv.warpAffine(imageP, M, (w, h), flags=cv.INTER_CUBIC, borderMode=cv.BORDER_REPLICATE)

    
    cv.waitKey(0)

    return image

def decode(out):
    # 0 means narrow, 1 means wide
    NARROW = "0"
    WIDE = "1"
    code11_widths = {
        "00110": "Stop/Start",
        "10001": "1",
        "01001": "2",
        "11000": "3",
        "00101": "4",
        "10100": "5",
        "01100": "6",
        "00011": "7",
        "10010": "8",
        "10000": "9",
        "00001": "0",
        "00100": "-",
    }

    # Get the average of each column in your image
    mean = out.mean(axis=0)
    
    # Set it to black or white based on its value
    mean[mean <= 127] = 1
    mean[mean > 128] = 0

    for i in range(len(mean)):
        if mean[i] == 0:
            mean[i] = 1
        else:
            break

    # Convert to string of pixels in order to loop over it
    pixels = ''.join(mean.astype(np.uint8).astype(str))

    # Need to figure out how many pixels represent a narrow bar
    narrow_bar_size = 0
    for pixel in pixels:
        if pixel == "1":
            narrow_bar_size += 1
        else:
            break

    wide_bar_size = narrow_bar_size * 2
    
    print(f"narrow: {narrow_bar_size}, wide: {wide_bar_size}")

    print(mean)

    digits = []
    pixel_index = 0
    current_digit_widths = ""
    skip_next = False

    while pixel_index < len(pixels):

        if skip_next:
            pixel_index += narrow_bar_size
            skip_next = False
            continue

        count = 1
        try:
            while pixels[pixel_index] == pixels[pixel_index + 1]:
                count += 1
                pixel_index += 1
        except:
            pass
        pixel_index += 1

        current_digit_widths += NARROW if narrow_bar_size - 1 <= count <= narrow_bar_size + 1 else WIDE

        if current_digit_widths in code11_widths:
            digits.append(code11_widths[current_digit_widths])
            current_digit_widths = ""
            skip_next = True  # Next iteration will be a separator, so skip it

    print(digits)

def blur_image(image):
    blur = cv.GaussianBlur(image, (5,5), cv.BORDER_DEFAULT)
    return image

def apply_median(image):
    # Apply Median Blur
    kernel = np.ones((3,1),np.float32)/3
    median_blurred_img = cv.filter2D(image,-1,kernel)
    return median_blurred_img
image_path = 'photos/09.jpg'
def crop_image_before_open(image):
        # Crop the barcode area using outlines
    barcode_segment, contour_visualization = isolate_barcode_section(image)
    visualize_image_with_title(cv.bitwise_not(barcode_segment), 'before open cropped Area')
    return barcode_segment
# Read the image
loaded_image = cv.imread(image_path)
if loaded_image is None:
    print(f"Failed to load the image: {image_path}")
else:
    isolated_barcode_image = extract_grayscale_regions_hsv(loaded_image, saturation_threshold=30)
    visualize_image_with_title(cv.cvtColor(isolated_barcode_image, cv.COLOR_BGR2RGB), 
                               "Isolated Grayscale Regions (HSV-Based)")
    detect_rotation(isolated_barcode_image)
    image_after_rotation=Handle_rotation(loaded_image)
    black_and_white = process_barcode_with_thresholding(image_after_rotation)
    #cropped_image=crop_image_before_open(black_and_white)
    #final_barcode = process_barcode_area_closing(cropped_image)
    #**************************************
    barcode = detectBarcode(255-black_and_white)
    visualize_image_with_title(barcode, 'Fioa')
    w, h = image_after_rotation.shape[:2]
    blured_barcode = cv.medianBlur(cv.blur(barcode,(1,h)),1)
    _, binary_barcode = cv.threshold(blured_barcode, 128, 255, cv.THRESH_BINARY)
    kernel = cv.getStructuringElement(cv.MORPH_RECT, (1, h)) 
    final = cv.morphologyEx(binary_barcode, cv.MORPH_OPEN, kernel)
    #*************************************
    decode(final)
```
